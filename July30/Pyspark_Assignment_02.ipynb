```python
#1.Create CSV file
import pandas as pd
import matplotlib.pyplot as plt

csv_data = """Month,Store,Sales
Jan,Store A,12000
Jan,Store B,15000
Feb,Store A,18000
Feb,Store B,17000
Mar,Store A,16000
Mar,Store B,19000
"""
# Saving to CSV
with open('sales_data.csv', 'w') as file:
    file.write(csv_data)
# Step 2: Read the CSV
df = pd.read_csv('sales_data.csv')
print("Full DataFrame:")
print(df)
```

```python
#2.Total sales per store
store_sales=df.groupby('Store')['Sales'].sum().reset_index()
print("\nTotal sales per store:")
print(store_sales)
#Identify store with highest sales
top_store=store_sales.loc[store_sales['Sales'].idxmax(),'Store']
print(f"\n Store with highest sales: {top_store}")
```

```python
#Monthly Sales Trend
monthly_sales=df.groupby('Month')['Sales'].sum().reset_index()
print("\nMonthly sales trend:")
print(monthly_sales)
```

```python
#Step5 Line Chart=Monthly sales by store
pivot_df=df.pivot(index='Month',columns='Store',values='Sales')
pivot_df.plot(kind='line',marker='o')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales by Store')
plt.grid(True)
plt.legend(title='Store')
plt.show()
```

```python
#5 Bonus
plt.figure(figsize=(8,5))
bars=plt.bar(store_sales['Store'],store_sales['Sales'],color=['skyblue','pink'])
plt.title('Total sales per store')
plt.xlabel('Store')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
#Adding data lables
for bar in bars:
  yval=bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,yval+500,round(yval,2),ha='center',va='bottom')
plt.show()
```
